```python
import pandas as pd
import os, re

files = os.listdir('../data/main_data/')

data = []

for month in files:

    for e in os.listdir(os.path.join('../data/main_data/',month)):
        searched = re.search('([a-z]).*', e)
        df = pd.read_csv(os.path.join('../data/main_data/', month, e))
        data.append(dict(
            month = month,
            location = e[searched.start(): searched.end()-4],
            **dict(df['Crime type'].value_counts())
        ))
```

```python
df = pd.DataFrame(data)
```
